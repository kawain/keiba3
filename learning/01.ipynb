## LabelEncoderで0からのint、pandasのカテゴリ型変換

```python
df1["race_type"].value_counts()
```

```python
# race_type を　ダ、芝、障に分ける
df1["race_type"] = df1["race_type"].str[0]
```

```python
df1["race_type"].value_counts()
```

```python
le = preprocessing.LabelEncoder()
```

```python
df1["race_type"] = le.fit_transform(df1["race_type"])
```

```python
df1["race_type"].value_counts()
```

```python
df1["race_type"] = df1["race_type"].astype('category')
```

```python
df1["race_course"] = le.fit_transform(df1["race_course"])
df1["race_course"] = df1["race_course"].astype('category')
```

```python
df1["race_distance"].value_counts()
```

```python
# 距離を区分け
df1["race_distance"] = pd.cut(df1["race_distance"], [0, 1399, 1799, 2199, 2799 ,5000])
```

```python
df1["race_distance"].value_counts()
```

```python
df1["race_distance"] = le.fit_transform(df1["race_distance"])
df1["race_distance"] = df1["race_distance"].astype('category')
```

```python
df1["race_distance"].value_counts()
```

```python
# race_weather race_state
df1["race_weather"] = le.fit_transform(df1["race_weather"])
df1["race_weather"] = df1["race_weather"].astype('category')

df1["race_state"] = le.fit_transform(df1["race_state"])
df1["race_state"] = df1["race_state"].astype('category')
```

```python
df1["race_weather"].value_counts()
```

```python
df1["race_state"].value_counts()
```

```python
df1["frame"].value_counts()
```

```python
df1["frame"] = df1["frame"].astype('category')
```

```python
df1["name_id"].value_counts()
```

```python
# gender
df1["gender"] = le.fit_transform(df1["gender"])
df1["gender"] = df1["gender"].astype('category')
```

```python
# jockey_id trainer dad_id mom_id grandpa_id
label_names = ["jockey_id", "trainer", "area", "dad_id", "mom_id", "grandpa_id"]

def matome_henkan(x):
    df1[x] = le.fit_transform(df1[x])
    df1[x] = df1[x].astype('category') 
    
```

```python
for v in label_names:
    matome_henkan(v)
```

```python
# order　を 0 から始める
df1["order"] = df1["order"] - 1
```

# 月と日のカラム

```python
df1["m"] = df1["race_date"].dt.month
df1["d"] = df1["race_date"].dt.day
```

```python
df1["m"].value_counts()
```

```python
df1["d"].value_counts()
```

```python
# 日を区分け
df1["d"] = pd.cut(df1["d"], [0, 19, 31])
```

```python
df1["d"].value_counts()
```

```python
for v in ["m", "d"]:
    matome_henkan(v)
```
