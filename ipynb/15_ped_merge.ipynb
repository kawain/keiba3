```python
head_df = pd.read_csv("../csv/head.csv")
body_df = pd.read_csv("../csv/body.csv", dtype={"jockey_id": object})
```

```python
print(head_df.shape)
print(body_df.shape)
print(ped_df.shape)
```

```python
# race_dateをdatetimeに変換
head_df["race_date"] = pd.to_datetime(head_df["race_date"], format="%Y年%m月%d日")
```

```python
df = pd.merge(head_df, body_df, on="race_id", how="outer")
```

```python
df.head(3)
```

```python
ped_df.info()
```

```python
ped_df = ped_df.astype({'my_id':'int64'})
```

```python
df = pd.merge(df, ped_df, left_on="name_id", right_on="my_id", how="inner")
```

```python
df.isnull().sum()
```

```python
agar_null = df.loc[df["agari"].isnull(), "race_id"]
```

```python
def agari_fillna(id):
    df.loc[(df["race_id"] == id) & (df["agari"].isnull()), "agari"] = df.loc[df["race_id"] == id]["agari"].max()+1
    
for v in agar_null:
    agari_fillna(v)
```

```python
df.isnull().sum()
```

```python
df.head(3)
```

```python
df["top_time"] = pd.to_datetime(df["top_time"], format="%M:%S.%f")
df["race_time"] = pd.to_datetime(df["race_time"], format="%M:%S.%f")
```

```python
df.head(2)
```

```python
# トップとの差
df["diff"] = (df["race_time"] - df["top_time"]).dt.total_seconds()
```

```python
df.head(1)
```

```python
# 不要カラム削除
df.drop(columns=[
    "top_time",
    "race_time",
    "id",
    "my_id"
], inplace=True)
```

```python
g = df.groupby("race_id")
g
```

```python
sad_max = g["speed","agari","diff"].max()
```

```python
sad_max.reset_index(inplace=True)
```

```python
sad_max = sad_max.rename(columns={"speed":"speed_max","agari":"agari_max","diff":"diff_max"})
```

```python
sad_max.head()
```

```python
df = pd.merge(df, sad_max, on="race_id")
```

```python
df.head(1)
```

```python
sad_min = g["speed","agari","diff"].min()
```

```python
sad_min.reset_index(inplace=True)
```

```python
sad_min = sad_min.rename(columns={"speed":"speed_min","agari":"agari_min","diff":"diff_min"})
```

```python
sad_min.head()
```

```python
df = pd.merge(df, sad_min, on="race_id")
```

```python
df.head(1)
```

```python
# "speed_min_max","agari_min_max","diff_min_max"
df["speed_min_max"] = (df["speed"]-df["speed_min"])/(df["speed_max"]-df["speed_min"])
df["agari_min_max"] = (df["agari"]-df["agari_min"])/(df["agari_max"]-df["agari_min"])
df["diff_min_max"] = (df["diff"]-df["diff_min"])/(df["diff_max"]-df["diff_min"])
```

```python
df.head()
```

```python
# ソート
df.sort_values(["race_date", "race_course", "race_number"], ascending=[False, True, True], inplace=True)
```

```python
# インデックスを振り直す
df.reset_index(inplace=True, drop=True)
```
